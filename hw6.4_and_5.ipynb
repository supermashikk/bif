```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar6/homework/TCGA-COAD.tsv', sep = '\t')
```

```python
data.head(5)
```

```python
df1 = data[['Gene','TCGA-A6-2671-11A']]
df2 = data[['Gene','TCGA-A6-2675-11A']]
```

```python
df1.sort_values('TCGA-A6-2671-11A', ascending = False)
```

```python
df2.sort_values('TCGA-A6-2675-11A', ascending = False)
```

```python
df1_100 = df1.sort_values('TCGA-A6-2671-11A', ascending = False)[0:100]
```

```python
df2_100 = df2.sort_values('TCGA-A6-2675-11A', ascending = False)[0:100]
```

```python
df1_100
```

```python
df1_100 = df1_100.set_index('Gene')
```

```python
df2_100 = df2_100.set_index('Gene')
```

```python
df1_100
```

```python
common = df1_100.join(df2_100)
```

```python
common
```

```python
common = common.dropna()
```
