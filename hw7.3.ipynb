```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from scipy.stats.mstats import gmean
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar7/homework/TCGA-COAD_cancer_normal.tsv', sep = '\t', index_col = 0)
```

```python
df = df.T
```

```python
df = df.loc[:, (df != 0).all(axis=0)]
```

```python
row = np.zeros(19013)
df.loc['Pseudo-sample'] = row
```

```python
df.loc['Pseudo-sample'] = gmean(df.iloc[0:10,:])
df = df.div(df.loc['Pseudo-sample'])
```

```python
df
```

```python
size_fac = df.median(axis = 1)
```

```python
size_fac
```

```python
df1 = df.T.melt (var_name = 'Sample', value_name = 'Correlation')
```

```python
df1
```

```python
df1 = df1[df1!= 'Pseudo-sample']
```

```python
df1 = df1.dropna()
```

```python
df1
```

```python
g = sns.barplot(data = df1, x = 'Sample', y = 'Correlation', estimator = np.median, color = 'orange')
g.set_xticklabels(g.get_xticklabels(),rotation=90)
plt.tight_layout()
```
