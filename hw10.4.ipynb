```python
from scipy.stats import t
from scipy.stats import norm
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

rv1 = t(1)


def graphs(a):
    
        df1 = pd.DataFrame({})
        for i in range (1,10000):
            df1[i] = rv1.rvs(size = a)

        df1.loc["Mean"] = df1.mean()
        print(df1)

        df2 = pd.DataFrame({
            "x": np.linspace(-3, 3, 10000)
        })
        
        df2["PDF"] = rv1.pdf(df2["x"])

        plt.tight_layout()
        
        sns.histplot(x = df1.loc["Mean"], color = "Pink", stat = "density", binrange = (-10,10))
        sns.lineplot(data = df2, x = "x", y = "PDF", color = "Green")


graphs(5)
```

```python
graphs(2)
```

```python
graphs(14)
```
